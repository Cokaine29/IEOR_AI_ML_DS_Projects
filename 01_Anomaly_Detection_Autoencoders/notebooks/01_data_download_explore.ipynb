# Phase 1: Data Acquisition and Exploration

In this notebook, we verify the downloaded CWRU bearing dataset, parse the metadata, and explore the time-domain vibration signals.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

## Load Processed Data
We load the windowed time-series segments and their corresponding labels.

```python
data_dir = "../data/processed"

X = np.load(os.path.join(data_dir, "X.npy"))
y = np.load(os.path.join(data_dir, "y.npy"))
meta = pd.read_csv(os.path.join(data_dir, "metadata.csv"))

print(f"Data shape (X): {X.shape}")
print(f"Labels shape (y): {y.shape}")
display(meta.head())
```

## Dataset Distribution

```python
# Count windows per fault type
fault_counts = meta['fault_type'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=fault_counts.index, y=fault_counts.values, hue=fault_counts.index)
plt.title("Number of Windows per Condition (Normal vs Faults)")
plt.ylabel("Count")
plt.show()
```

## Visualizing Vibration Signals
Let's compare a normal vibration signal with an inner race fault and a ball fault.

```python
def plot_signal(ax, data, title, color):
    ax.plot(data, color=color, linewidth=0.8)
    ax.set_title(title)
    ax.set_ylim([-3.5, 3.5])
    ax.set_xlabel("Time Steps")
    ax.set_ylabel("Acceleration (g)")

fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

# Get one normal window
idx_normal = meta[meta['fault_type'] == 'normal'].index[0]
plot_signal(axes[0], X[idx_normal], "Normal Vibration Signal", "blue")

# Get one inner race fault window
idx_inner = meta[meta['fault_type'] == 'inner'].index[0]
plot_signal(axes[1], X[idx_inner], "Inner Race Fault Signal", "red")

# Get one ball fault window
idx_ball = meta[meta['fault_type'] == 'ball'].index[0]
plot_signal(axes[2], X[idx_ball], "Ball Fault Signal", "orange")

plt.tight_layout()
plt.show()
```
