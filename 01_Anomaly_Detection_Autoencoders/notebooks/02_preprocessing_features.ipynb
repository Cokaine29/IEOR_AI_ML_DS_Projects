# Phase 2: Feature Engineering

In this notebook, we verify the extracted time-domain and frequency-domain features that will be used to train our classical baseline models (Isolation Forest and One-Class SVM).

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

## Load Processed Features and Metadata

```python
data_dir = "../data/processed"

# Load features and metadata
features_df = pd.read_csv(os.path.join(data_dir, "features.csv"))
meta_df = pd.read_csv(os.path.join(data_dir, "metadata.csv"))

# Merge them for easy plotting
df = pd.concat([features_df, meta_df[['fault_type', 'severity']]], axis=1)

print(f"Feature matrix shape: {features_df.shape}")
display(df.head())
```

## Feature Discriminability

For classical anomaly detection to work, the features extracted from 'normal' windows must look statistically different from those extracted from 'fault' windows.

Let's look at **RMS (Root Mean Square)**, which measures overall vibration energy, and **Kurtosis**, which measures the "peakedness" (impulses) of the signal.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# RMS Boxplot
sns.boxplot(data=df, x='fault_type', y='rms', ax=axes[0], hue='fault_type')
axes[0].set_title("Distribution of RMS (Overall Energy)")
axes[0].set_ylabel("RMS Acceleration")

# Kurtosis Boxplot
sns.boxplot(data=df, x='fault_type', y='kurtosis', ax=axes[1], hue='fault_type')
axes[1].set_title("Distribution of Kurtosis (Impulsiveness)")
axes[1].set_ylabel("Kurtosis")

plt.tight_layout()
plt.show()
```

## Frequency Domain Features
Let's look at **Spectral Entropy**. Normal bearings have a wider, flatter noise spectrum (higher entropy). Faulty bearings concentrate energy at specific fault frequencies (lower entropy).

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='fault_type', y='spectral_entropy', hue='fault_type')
plt.title("Distribution of Spectral Entropy")
plt.ylabel("Spectral Entropy")
plt.show()
```
